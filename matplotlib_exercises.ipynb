```python
import matplotlib.pyplot as plt 
from numpy import linspace

from math import sqrt
```

1. Use matplotlib to plot the following equation:

y=x^2âˆ’x+2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = list(range(-10, 10))
y = [n**2 - n+2 for n in x]
plt.plot(x,y)
plt.plot(0,0, marker = 'o')
plt.title('A Quadratic Distribution, $y = x^2 - x+2$', c='white')
plt.xlabel('$x$', c='white')
plt.ylabel('$x^2- x+2$', c='white')
plt.text(0.25, 0, '(0, 0)', fontsize=20, color='red')
plt.tick_params(axis='x', colors='w')
plt.tick_params(axis='y', colors='w')
plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.



```python
# y=âˆšx
x1 = list(range(0, 15))
y1 = [sqrt(n) for n in x1]
plt.plot(x1,y1, c = 'red', label = '$y=âˆšx$')
plt.plot(0,0, marker = 'o')

#y=x^3
x2 = list (linspace(-10,10, num = 1000))
y2 = [n**3 for n in x2]
plt.plot(x2,y2, c = 'blue', label = '$y=x^3$')

#y=2^x
x3 = list (linspace(-10,10, num = 1000))
y3 = [2**n for n in x3]
plt.plot(x3,y3, c = 'black', label = '$y=2^x$')

#y=1/(x+1)
x4 = list (linspace(-1,10, num = 10000))
x4.remove(-1.0)
y4 = [1/(n+1) for n in x4]
plt.plot(x4,y4, c = 'green', label = '$y=1/(x+1)$')
x4 = list (linspace(-10,-1, num = 10000))
x4.remove(-1.0)
y4 = [1/(n+1) for n in x4]
plt.plot(x4,y4, c = 'green')



font1 = {'family':'serif','color':'white','size':20}
plt.title('Combination of functions',  fontdict= font1)
plt.xlabel("x", fontdict= font1)
plt.ylabel("y", fontdict= font1)
plt.tick_params(axis='x', colors='w')
plt.tick_params(axis='y', colors='w')
plt.legend(loc='lower right')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.savefig('combinations-of-functions')
plt.show()
```
