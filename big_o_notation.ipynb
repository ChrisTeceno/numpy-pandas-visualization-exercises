Curves to graph

y=0n+1 and label the curve "O(1)"

y=log(n) and label the curve "O(log n)"

y=n and label the curve "O(n)"

y=n∗log(n) and label it "O(n log n)"

y=n^2 and label it "O(n^2)"

y=2^n and label it "O(2^n)"

y=n! and label it "O(n!)"

y=n^n and label it "O(n^n)"

```python
import matplotlib.pyplot as plt 
from numpy import linspace

from math import sqrt, log, factorial
```

```python
#y=0n+1 and label the curve "O(1)"
x1 = list(range(0, 1000))
y1 = [0*n+1 for n in x1]
plt.plot(x1,y1, c = 'red', label = '$O(1)$')
plt.plot(0,0, marker = 'o')

#y=log(n) and label the curve "O(log n)"
x2 = list (linspace(0,1000, num = 100))
x2.remove(0.0)
y2 = [log(n) for n in x2]
plt.plot(x2,y2, c = 'g', label = '$O(log n)$')

#y=n and label the curve "O(n)"
x3 = list (linspace(0,100, num = 100))
x3.remove(0.0)
y3 = [n for n in x3]
plt.plot(x3,y3, c = 'b', label = '$O(n)$')

#y=n∗log(n) and label it "O(n log n)"
x4 = list (linspace(0,100, num = 1000))
x4.remove(0.0)
y4 = [n*log(n) for n in x4]
plt.plot(x4,y4, c = 'c', label = '$O(n log n)$')

# y=n^2 and label it "O(n^2)"
x5 = list (linspace(0,100, num = 1000))
x5.remove(0.0)
y5 = [n**2 for n in x5]
plt.plot(x5,y5, c = 'm', label = '$O(n^2)$')

# y=2^n and label it "O(2^n)"
x6 = list (linspace(0,100, num = 1000))
x6.remove(0.0)
y6 = [2**n for n in x6]
plt.plot(x6,y6, c = 'y', label = '$O(2^n)$')

# y=n! and label it "O(n!)"
x7 = list(range(0,100))
x7.remove(0.0)
y7 = [factorial(n) for n in x7]
plt.plot(x7,y7, c = 'k', label = '$O(n!)$')

# y=n^n and label it "O(n^n)"
x8 = list (linspace(0,100, num = 1000))
x8.remove(0.0)
y8 = [n**n for n in x8]
plt.plot(x8,y8, c = 'firebrick', label = '$O(n^n)$')

font1 = {'family':'serif','color':'white','size':20}
plt.title('Big O Notation',  fontdict= font1)
plt.xlabel("Elements", fontdict= font1)
plt.ylabel("Operations", fontdict= font1)
plt.tick_params(axis='x', colors='w')
plt.tick_params(axis='y', colors='w')
plt.legend(loc='upper right')
plt.xlim(0, 100)
plt.ylim(0, 200)
plt.savefig('big-o')
plt.show()
```
